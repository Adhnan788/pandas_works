```python
df.rename(columns={"Customer ID":"customer_id","Gender":"gender","Age":"age","City":"city","Membership Type":"membership_type","Total Spend":"total_spend","Items Purchased":"item_purchased","Average Rating":"average_rating","Discount Applied ":"discount_applied","Days Since Last Purchase":"days_since_last_purchase","Satisfaction Level":"satisfication_level"},inplace=True)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()

```

```python
condition=df["average_rating"].apply(lambda x: "Unsatisfied" if x<3.5 else "Neutral" if x>3.5 and x<4.0 else "Satisfied"  )
df["satisfication_level"].fillna(condition ,inplace=True)
```

```python
import matplotlib.pyplot as plt
gender_count=df["gender"].value_counts()
plt.plot(gender_count,marker="o",linestyle="none")
plt.title("gender count")
plt.xlabel("gender")
plt.ylabel("count")
plt.show()
```

```python
#membership wise total_spent sum
sum_spent=df.groupby("membership_type")["total_spend"].sum()
```
